```python
import os
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python


IMAGE_PATH = "../datasets/earrings.jpg"
RESULTS_DIR = "../results/Q2"

os.makedirs(RESULTS_DIR, exist_ok=True)

im = cv.imread(IMAGE_PATH)
assert im is not None, "Image not found. Check IMAGE_PATH."

im_rgb = cv.cvtColor(im, cv.COLOR_BGR2RGB)

print("Image shape:", im.shape)
print("Image dtype:", im.dtype)
```

The real size of the earrings can be estimated using the pinhole camera scale relationship:

real size = image size on sensor × object distance / focal length

The image size on the sensor is:

image size on sensor = number of pixels × pixel size

Given:
focal length = 8 mm
pixel size = 2.2 µm = 0.0022 mm
object distance = 720 mm

Therefore:

real size per pixel = 0.0022 × 720 / 8 = 0.198 mm/pixel

```python
FOCAL_LENGTH_MM = 8
PIXEL_SIZE_MM = 2.2 / 1000   
DISTANCE_MM = 720

MM_PER_PIXEL = PIXEL_SIZE_MM * DISTANCE_MM / FOCAL_LENGTH_MM

print(f"Scale: {MM_PER_PIXEL:.3f} mm/pixel")
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(im_rgb)
plt.title("Earrings image with pixel coordinates")
plt.xlabel("x pixel")
plt.ylabel("y pixel")
plt.grid()
plt.show()
```

```python


left_hoop_bbox = (250, 405, 470, 690)
right_hoop_bbox = (555, 405, 775, 690)

left_full_bbox = (240, 285, 480, 700)
right_full_bbox = (545, 285, 785, 700)
```

```python
def bbox_measurement_mm(bbox, mm_per_pixel):
    x_min, y_min, x_max, y_max = bbox
    
    width_px = x_max - x_min
    height_px = y_max - y_min
    
    width_mm = width_px * mm_per_pixel
    height_mm = height_px * mm_per_pixel
    
    return width_px, height_px, width_mm, height_mm


measurements = {
    "Left hoop outer size": bbox_measurement_mm(left_hoop_bbox, MM_PER_PIXEL),
    "Right hoop outer size": bbox_measurement_mm(right_hoop_bbox, MM_PER_PIXEL),
    "Left full size including hook": bbox_measurement_mm(left_full_bbox, MM_PER_PIXEL),
    "Right full size including hook": bbox_measurement_mm(right_full_bbox, MM_PER_PIXEL),
}

for name, (w_px, h_px, w_mm, h_mm) in measurements.items():
    print(name)
    print(f"  width  = {w_px:.1f} px = {w_mm:.2f} mm")
    print(f"  height = {h_px:.1f} px = {h_mm:.2f} mm")
    print()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(im_rgb)

def draw_bbox(ax, bbox, label):
    x_min, y_min, x_max, y_max = bbox
    width = x_max - x_min
    height = y_max - y_min
    
    rect = plt.Rectangle(
        (x_min, y_min),
        width,
        height,
        fill=False,
        linewidth=2
    )
    ax.add_patch(rect)
    ax.text(x_min, y_min - 10, label, fontsize=10)

draw_bbox(ax, left_hoop_bbox, "Left hoop")
draw_bbox(ax, right_hoop_bbox, "Right hoop")
draw_bbox(ax, left_full_bbox, "Left full")
draw_bbox(ax, right_full_bbox, "Right full")

ax.set_title("Q2: Manual Pixel Measurements of Earrings")
ax.set_xlabel("x pixel")
ax.set_ylabel("y pixel")
ax.grid(True)

plt.tight_layout()
plt.savefig(f"{RESULTS_DIR}/q2_earring_measurement.png", dpi=300)
plt.show()
```

```python
with open(f"{RESULTS_DIR}/q2_measurements.txt", "w") as f:
    f.write("Q2: Earring size estimation\n")
    f.write(f"Scale = {MM_PER_PIXEL:.3f} mm/pixel\n\n")
    
    for name, (w_px, h_px, w_mm, h_mm) in measurements.items():
        f.write(f"{name}\n")
        f.write(f"width  = {w_px:.1f} px = {w_mm:.2f} mm\n")
        f.write(f"height = {h_px:.1f} px = {h_mm:.2f} mm\n\n")

print("Saved Q2 measurements.")
```
